```python
df_comment = df_comment.drop_duplicates(subset=["reply"])
```

```python
import numpy as np

df_comment = df_comment.loc[df_comment["reply"] != ""]
df_comment["count1"] = df_comment["comment"].map(lambda x: len(x))
df_comment["stat1"] = df_comment["comment"].map(lambda x: np.log(len(x)))
df_comment["count2"] = df_comment["reply"].map(lambda x: len(x))
df_comment["stat2"] = df_comment["reply"].map(lambda x: np.log(len(x)))
# mean = df_comment["stat"].mean()
# std = df_comment["stat"].std()

# a = 1
# len(df_comment.loc[(df_comment["stat"] > mean - a*std) & (df_comment["stat"] < mean + a*std)]) / len(df_comment)
```
